#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
def manual_std(data_series):
    return round(((((data_series - np.mean(data_series))**2).sum()) / len(data_series)) ** 0.5,5)
````

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`

````python
{i:manual_std(test_data[i]) for i in test_data.select_dtypes(include=np.number).columns}
````

Do your manually calculated standard deviations match up with the output from pandas `describe`? What about numpy's `std` method?

````python
# Test of function
print('std[sat_math] from .describe(): {}'.format(round(test_data['sat_math_2017'].describe()['std'],5)))
print('std[sat_math] from .manual_std():{}'.format(manual_std(test_data['sat_math_2017'])))
print('std[sat_math] from .np.std(): {}'.format(round(np.std(test_data['sat_math_2017']),5)))
````

<details><summary>Answer</summary>
Python.describe() uses an unbiased estimator (N-1) while Numpy & manual_std use N. Hence resulting in the difference.

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. While you can use pandas plotting or Seaborn here, matplotlib gives you greater control over all aspects of your plots.

[Helpful Link for Plotting Multiple Figures](https://matplotlib.org/users/pyplot_tutorial.html#working-with-multiple-figures-and-axes)

Here's some starter code:

````python
def subplot_histograms(dataframe, list_of_columns, list_of_titles, \
                       list_of_xlabels,x_arange,  figsize_tuple, fig_title):
    nrow = int(np.ceil(len(list_of_columns)/2)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrow, ncols=2, figsize=figsize_tuple) # You'll want to specify your figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    fig.suptitle(fig_title, fontsize=16)
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column], bins=x_arange) # feel free to add more settings
        ax[i].set_xticks(x_arange)
        ax[i].set_xlabel(list_of_xlabels[i])
        ax[i].set_ylabel(list_of_titles[i])
        hist,bin_edges = np.histogram(dataframe[column], bins=x_arange)
        for val in range(len(hist)):
            if hist[val] != 0:
                xpos = (bin_edges[val] + bin_edges[val+1]) / 2
                ax[i].text(xpos,hist[val],hist[val], ha='center', fontsize=10,)
````

#### Plot and interpret boxplots

For each numeric variable in the dataframe create a boxplot using Seaborn. Boxplots demonstrate central tendency and spread in variables. In a certain sense, these are somewhat redundant with histograms, but you may be better able to identify clear outliers or differences in IQR, etc.

Multiple values can be plotted to a single boxplot as long as they are of the same relative scale (meaning they have similar min/max values).

Each boxplot should:
- Only include variables of a similar scale
- Have clear labels for each variable
- Have appropriate titles and labels

````python
def plot_boxplt(dataframe, list_of_columns, ax_title, ax_ylabel, ax_xlabel, orient='v', figsize=(10,10)):
    fig, ax = plt.subplots(1, 1, figsize=figsize)
    ax = sns.boxplot(data=dataframe[list_of_columns], orient=orient)
    ax.set_title(ax_title)
    ax.set_ylabel(ax_ylabel)
    ax.set_xlabel(ax_xlabel)
    ax.grid(color='blue', linestyle='-.', linewidth=1)
````

#### Statistical Evaluation of Distributions 

**If you feel it's appropriate**, using methods we discussed in class, run hypothesis tests to compare variables of interest in our dataset. 

<b style="color:purple">H_null = students with higher SAT math score is NOT BETTER than those with lower ACT math score, or vice versa.</b>

<b style="color:purple">H_one = student with higher SAT math score is BETTER than those with lower ACT math score, or vice versa.</b>

````python

## rescale SAT_math to the same scale as ACT
sat_math = test_data['sat_math_2017']*36/800
act_math = test_data['act_math_2017']
print(f'p-value = {stats.ttest_ind(sat_math, act_math, equal_var=False)[1]}')
````

<details><summary>Answer</summary>
Given the p-value is less than the threshold p-value of 0.05 to void the null hypothesis. We can infer statistically that students with higher SAT math score is better tha nthose with lower ACT math score, or vice versa.
</details>
