### User Custom Functions

```python
import sklearn.metrics as metrics
import numpy as np

def lord_of_the_metrics(y_true, y_pred, p):
    '''p = Number of predictors '''
    mse = metrics.mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    msle = metrics.mean_squared_log_error(y_true, y_pred)
    mae = metrics.median_absolute_error(y_true, y_pred)
    r2 = metrics.r2_score(y_true, y_pred)
    r2a = r2_adj(y_true, y_pred, p)
    
    print('Mean squared error      = ', mse)
    print('Root mean squared error = ', rmse)
    print('Mean squared log error  = ', msle)
    print('Median absolute error   = ', mae)
    print('R^2                     = ', r2)
    print('Adjusted R^2            = ', r2a)
```

```python
def r2_adj(y_true, y_preds, p):
    n = len(y_true)
    y_mean = np.mean(y_true)
    numerator = np.sum(np.square(y_true - y_preds)) / (n - p - 1)
    denominator = np.sum(np.square(y_true - y_mean)) / (n - 1)
    return 1 - numerator / denominator
```

```python

def internal_print(name_tag, metric_tag, train_score, test_score, cv_s_mean=None, n_title=False):
    if n_title == True: print(('-'*60 + ' {} Model Metrics ' + '-' *60).format(name_tag))
    if cv_s_mean != None: print(f'{metric_tag} of CV mean ({name_tag}): {cv_s_mean: ,.5f}')
    print(f'{metric_tag} of Training ({name_tag}): {train_score: ,.5f}')
    print(f'{metric_tag} of Test ({name_tag}): {test_score: ,.5f}')
    print('\n')
    return None
```

```python
def lasso_elim(X_train, X_test, y_train, y_test,features_list,ls_alpha=100, lasso_iter=1000):
    '''
    Function: To generate a list of features to eliminate based on lasso regularization



    return a dataframe of coefficent of features. Features = index
    '''

    # 1. Instantiate StandardScaler and scaled data
    ss = StandardScaler()
    X_train_scaled = ss.fit_transform(X_train)
    X_test_scaled = ss.transform(X_test)
    
    # 2. Instantiate Models, Cross-Val and Fitting
    lasso_cv = LassoCV(n_alphas = ls_alpha, max_iter=lasso_iter)
    
    lasso_cv.fit(X_train_scaled, y_train)

    # 3. Execute Cross_val_scoring
    lasso_cv_scores = cross_val_score(lasso_cv, X_train_scaled,y_train, cv=5)
    lasso_cv_mse_scores = -cross_val_score(lasso_cv, X_train_scaled,y_train, cv=5,scoring ='neg_mean_squared_error')
    
    # 4. Printing result

    ## Lasso Regularisation
    internal_print('Lasso', 'R^2',lasso_cv.score(X_train_scaled,y_train), lasso_cv.score(X_test_scaled,y_test),lasso_cv_scores.mean(), n_title=True)
    internal_print('Lasso', 'Adj-R^2',r2_adj(y_train, lasso_cv.predict(X_train_scaled), len(X_train.columns)), \
                   r2_adj(y_test, lasso_cv.predict(X_test_scaled), len(X_test.columns)))
    internal_print('Lasso', 'RMSE',np.sqrt(metrics.mean_squared_error(y_train, lasso_cv.predict(X_train_scaled))),\
                   np.sqrt(metrics.mean_squared_error(y_test, lasso_cv.predict(X_test_scaled))),np.sqrt(lasso_cv_mse_scores.mean()))

    # 5. Print out Lasso Feature Mapping
    pd.Series(lasso_cv.coef_, index=features_list).plot.barh(figsize=(15,20))
    rund_fea = X_train.columns[lasso_cv.coef_==0]
    print(f'Number of redundant features to drop: {len(rund_fea)}')
    
    return lasso_cv.coef_
```

```python
def Modeling(X_train, X_test, y_train, y_test,features_list,ls_alpha=100, lasso_iter=1000, ri_alpha=(0.1,10,100)):
    '''
    Function: To scale train, test data and fit into linear regression, ridge regularization & lasso regularization and plot chart of y vs y-pred



    return None
    '''

    # 1. Instantiate StandardScaler and scaled data
    ss = StandardScaler()
    X_train_scaled = ss.fit_transform(X_train)
    X_test_scaled = ss.transform(X_test)
    
    # 2. Instantiate Models, Cross-Val and Fitting
    lr = LinearRegression()
    lasso_cv = LassoCV(n_alphas = ls_alpha, max_iter=lasso_iter)
    ridge_cv = RidgeCV(alphas=ri_alpha)
    
    lr.fit(X_train_scaled, y_train)
    lasso_cv.fit(X_train_scaled, y_train)
    ridge_cv.fit(X_train_scaled, y_train)

    # 3. Execute Cross_val_scoring
    lr_cv_scores = cross_val_score(lr, X_train_scaled,y_train, cv=5)
    lr_cv_mse_scores = -cross_val_score(lr, X_train_scaled,y_train, cv=5,scoring ='neg_mean_squared_error')
    ridge_cv_scores = cross_val_score(ridge_cv,X_train_scaled,y_train,cv=5)
    ridge_cv_mse_scores = -cross_val_score(ridge_cv,X_train_scaled,y_train,cv=5, scoring ='neg_mean_squared_error')
    lasso_cv_scores = cross_val_score(lasso_cv, X_train_scaled,y_train, cv=5)
    lasso_cv_mse_scores = -cross_val_score(lasso_cv, X_train_scaled,y_train, cv=5,scoring ='neg_mean_squared_error')
    
    # 4. Printing result
    
    ## Linear Regression
    internal_print('LR', 'R^2',lr.score(X_train_scaled,y_train), lr.score(X_test_scaled,y_test),lr_cv_scores.mean(), n_title=True)
    
    internal_print('LR', 'Adj-R^2',r2_adj(y_train, lr.predict(X_train_scaled), len(X_train.columns)), \
                   r2_adj(y_test, lr.predict(X_test_scaled), len(X_test.columns)))
    internal_print('LR', 'RMSE',np.sqrt(metrics.mean_squared_error(y_train, lr.predict(X_train_scaled))), \
                   np.sqrt(metrics.mean_squared_error(y_test, lr.predict(X_test_scaled))),np.sqrt(lr_cv_mse_scores.mean()))
    
    ## Ridge Regularisation
    internal_print('Ridge', 'R^2',ridge_cv.score(X_train_scaled,y_train), ridge_cv.score(X_test_scaled,y_test),ridge_cv_scores.mean(), n_title=True)
    internal_print('Ridge', 'Adj-R^2',r2_adj(y_train, ridge_cv.predict(X_train_scaled), len(X_train.columns)), \
                   r2_adj(y_test, ridge_cv.predict(X_test_scaled), len(X_test.columns)))
    internal_print('Ridge', 'RMSE',np.sqrt(metrics.mean_squared_error(y_train, ridge_cv.predict(X_train_scaled))), \
                   np.sqrt(metrics.mean_squared_error(y_test, ridge_cv.predict(X_test_scaled))),np.sqrt(ridge_cv_mse_scores.mean()))

    ## Lasso Regularisation
    internal_print('Lasso', 'R^2',lasso_cv.score(X_train_scaled,y_train), lasso_cv.score(X_test_scaled,y_test),lasso_cv_scores.mean(), n_title=True)
    internal_print('Lasso', 'Adj-R^2',r2_adj(y_train, lasso_cv.predict(X_train_scaled), len(X_train.columns)), \
                   r2_adj(y_test, lasso_cv.predict(X_test_scaled), len(X_test.columns)))
    internal_print('Lasso', 'RMSE',np.sqrt(metrics.mean_squared_error(y_train, lasso_cv.predict(X_train_scaled))),\
                   np.sqrt(metrics.mean_squared_error(y_test, lasso_cv.predict(X_test_scaled))),np.sqrt(lasso_cv_mse_scores.mean()))

    
    # 5. Plotting of charts
    ##Plot Setting
    fig = plt.figure(figsize=(10,15))
    gs = fig.add_gridspec(3, 1,top = 0.85, hspace= 0.3, wspace=0.4)
    plt.title('3 Model plots', fontsize=20, pad=10)
    plt.axis('off')
    
    # Linear Regression Plot
    f_ax1 = fig.add_subplot(gs[0, 0])
    sns.scatterplot(x=lr.predict(X_test_scaled), y=y_test, alpha=0.5)
    sns.lineplot(x=[0,np.max(y_test)], y=[0,np.max(y_test)], color='red', label='Act = Pred')
    f_ax1.set_title('Linear Regression Plot of Actual vs Predict on validation data')
    f_ax1.set_xlabel('Predicted saleprice')
    f_ax1.set_ylabel('Actual saleprice')
    f_ax1.legend()
 
    # Ridge Regularisation
    f_ax2 = fig.add_subplot(gs[1, 0])
    sns.scatterplot(x=ridge_cv.predict(X_test_scaled), y=y_test, alpha=0.5)
    sns.lineplot(x=[0,np.max(y_test)], y=[0,np.max(y_test)], color='red', label='Act = Pred')
    f_ax2.set_title('Ridge Regularization Plot of Actual vs Predict on validation data')
    f_ax2.set_xlabel('Predicted saleprice')
    f_ax2.set_ylabel('Actual saleprice')
    f_ax2.legend()
    
    # Lasso Regularisation
    f_ax3 = fig.add_subplot(gs[2, 0])
    sns.scatterplot(x=lasso_cv.predict(X_test_scaled), y=y_test, alpha=0.5)
    sns.lineplot(x=[0,np.max(y_test)], y=[0,np.max(y_test)], color='red', label='Act = Pred')
    f_ax3.set_title('Lasso Regularization Plot of Actual vs Predict on validation data')
    f_ax3.set_xlabel('Predicted saleprice')
    f_ax3.set_ylabel('Actual saleprice')
    f_ax3.legend()
    
```

```python
# To print mse from dataframe
def print_mse(dataframe, var_1, var_2):
    print(f'Shape of dataframe: {dataframe.shape}')
    print(f"Base Mean Squared Error: {metrics.mean_squared_error(dataframe[var_1], dataframe[var_2]):,.3f}")
    print(f"Base Root Mean Squared Error: {np.sqrt(metrics.mean_squared_error(dataframe[var_1], dataframe[var_2])):,.3f}")
    return None
```

```python
# Function to create polynomial - interaction term only and merge with dataframe
def poly_term(dataframe, poly_var_list, degree=None, concat=False):
    # Create a dataframe of given list
    df = dataframe[poly_var_list]
    
    if degree == None:
        degree = len(poly_var_list)
    else:
        degree = 2
    
    # Instantiate & transform
    poly = PolynomialFeatures(degree = degree,interaction_only=True ,include_bias=False)
    pf = poly.fit_transform(df)
    df2 = pd.DataFrame(pf, columns=poly.get_feature_names(poly_var_list))
    # Only poly features only
    df2.drop(columns=poly_var_list, inplace=True)
    if concat == False:
        return df2
    else:
        dataframe = pd.concat([dataframe,df2], axis=1)
        return dataframe
```

- Out of the 3 models, Ridge performed the best with a higher Adj-R^2 & lower RMSE.
