## Custom Functions

```python
def Modeling(X_train, X_test, y_train, y_test,features_list,ls_alpha=100, ri_alpha=(0.1,10,100)):
    '''
    Function: To scale train, test data and fit into linear regression, ridge regularization & lasso regularization



    return a dataframe of coefficent of features. Features = index
    '''
    
    # 1. Instantiate StandardScaler and scaled data
    ss = StandardScaler()
    X_train_scaled = ss.fit_transform(X_train)
    X_test_scaled = ss.transform(X_test)
    
    # 2. Instantiate Models, Cross-Val and Fitting
    lr = LinearRegression()
    lasso_cv = LassoCV(n_alphas = ls_alpha)
    ridge_cv = RidgeCV(alphas=ri_alpha)
    
    lr.fit(X_train_scaled, y_train)
    lasso_cv.fit(X_train_scaled, y_train)
    ridge_cv.fit(X_train_scaled, y_train)

    
    # 3. Execute Cross_val_scoring
    lr_cv_scores = cross_val_score(lr, X_train_scaled,y_train, cv=5)
    lr_cv_mse_scores = -cross_val_score(lr, X_train_scaled,y_train, cv=5,scoring ='neg_mean_squared_error')
    ridge_cv_scores = cross_val_score(ridge_cv,X_train_scaled,y_train,cv=5)
    ridge_cv_mse_scores = -cross_val_score(ridge_cv,X_train_scaled,y_train,cv=5, scoring ='neg_mean_squared_error')
    lasso_cv_scores = cross_val_score(lasso_cv, X_train_scaled,y_train, cv=5)
    lasso_cv_mse_scores = -cross_val_score(lasso_cv, X_train_scaled,y_train, cv=5,scoring ='neg_mean_squared_error')
    
    # 4. Printing result
    print('Linear Regression R^2 Value & MSE:')
    
    print(f'R^2 of Training (LR): {lr.score(X_train_scaled,y_train): ,.5f}')
    print(f'R^2 of Test (LR): {lr.score(X_test_scaled,y_test): ,.5f}')
    print(f'R^2 of Estimate of Testing (LR): {lr_cv_scores.mean():,.5f}')
    print('-' * 50)
    
    print(f'MSE of Training (LR): {metrics.mean_squared_error(y_train, lr.predict(X_train_scaled)): ,.5f}')
    print(f'MSE of Test (LR): {metrics.mean_squared_error(y_test, lr.predict(X_test_scaled)): ,.5f}')
    print(f'MSE of Estimate of Testing (LR): {lr_cv_mse_scores.mean(): ,.5f}')
    print('\n')
    
    print('Ridge Regularisation R^2 Value & MSE:')
    
    print(f'R^2 of Training (Ridge): {ridge_cv.score(X_train_scaled,y_train): ,.5f}')
    print(f'R^2 of Test (Ridge): {ridge_cv.score(X_test_scaled,y_test): ,.5f}')
    print(f'R^2 of Estimate of Testing (Ridge): {ridge_cv_scores.mean():,.5f}')
    print('-' * 50)
    
    print(f'MSE of Training (Ridge): {metrics.mean_squared_error(y_train, ridge_cv.predict(X_train_scaled)): ,.5f}')
    print(f'MSE of Test (Ridge): {metrics.mean_squared_error(y_test, ridge_cv.predict(X_test_scaled)): ,.5f}')
    print(f'MSE of Estimate of Testing (Ridge): {ridge_cv_mse_scores.mean(): ,.5f}')
    print('\n')
    
    
    print('Lasso Regularisation R^2 Value & MSE:')
    
    print(f'R^2 of Training (Lasso): {lasso_cv.score(X_train_scaled,y_train)}')
    print(f'R^2 of Test (Lasso): {lasso_cv.score(X_test_scaled,y_test): ,.5f}')
    print(f'R^2 of Estimate of Testing (Lasso): {lasso_cv_scores.mean():,.5f}')
    print('-' * 50)
    
    print(f'MSE of Training (Lasso): {metrics.mean_squared_error(y_train, lasso_cv.predict(X_train_scaled)): ,.5f}')
    print(f'MSE of Test (Lasso): {metrics.mean_squared_error(y_test, lasso_cv.predict(X_test_scaled)): ,.5f}')
    print(f'MSE of Estimate of Testing (Lasso): {lasso_cv_mse_scores.mean(): ,.5f}')
    
    # 5. Print out Lasso Feature Mapping
    pd.Series(lasso_cv.coef_, index=features_list).plot.barh(figsize=(15,20))
    
    # 6. Create a list of redundant features
    lasso_cv_coff = pd.DataFrame(lasso_cv.coef_,index=features_list, columns=['ls_cv_coff'])
    rund_fea = [i for i in lasso_cv_coff[lasso_cv_coff['ls_cv_coff']==0].index]
    print(f'Number of redundant features to drop: {len(rund_fea)}')
    
    return lasso_cv_coff

```

```python
def dic_insert (coeff_dataframe, delimit_pos, target_dict, target_dict_key, target_dataframe, target_map=False):
    coeff_dataframe['ls_cv_coff']= coeff_dataframe['ls_cv_coff'].map(lambda x: 'good' if x> 0 else ('neutral' if x ==0 else 'bad'))
    target_dict[target_dict_key] = {i.split('_')[delimit_pos]: coeff_dataframe.loc[i,'ls_cv_coff'] for i in coeff_dataframe.index}
    
    if target_map == True:
        target_dataframe[target_dict_key]=target_dataframe[target_dict_key].map(target_dict[target_dict_key])
```

```python
def dummy_encoder(dataframe, x_var_list, y_var):
    
    # Create dummy variable
    df_list = x_var_list.copy()
    df_list.append(y_var)
    dummy_df = pd.get_dummies(data=dataframe[df_list], columns=x_var_list)
    
    # Generate Matrix
    X = dummy_df.loc[:, dummy_df.columns !=y_var]
    y = dummy_df[y_var]
    
    # Create Train, test, split
    X_train, X_test, y_train, y_test =  train_test_split(X,y,random_state=42)
    
    return (X_train, X_test, y_train, y_test)
```
